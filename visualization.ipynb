```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("darkgrid")

# Load the results
asml_results_path = r'C:\Users\vagel\Desktop\Forecasting Realized Volatility\asml_realized_volatility.csv'
bmw_results_path = r'C:\Users\vagel\Desktop\Forecasting Realized Volatility\bmw_realized_volatility.csv'



# Read CSV files
asml_df = pd.read_csv(asml_results_path)
bmw_df = pd.read_csv(bmw_results_path)

# Convert 'date' column to datetime
asml_df["Date"] = pd.to_datetime(asml_df["Date"])
bmw_df["Date"] = pd.to_datetime(bmw_df["Date"])

# Set 'date' as the index
asml_df.set_index("Date", inplace=True)
bmw_df.set_index("Date", inplace=True)

# Plot ASML and BMW realized volatilities
plt.figure(figsize=(12, 6))
plt.plot(asml_df.index, asml_df["realized_volatility"], label="ASML", marker="o")
plt.plot(bmw_df.index, bmw_df["realized_volatility"], label="BMW", marker="s")

# Title and labels
plt.title("Daily Realized Volatility of ASML & BMW", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Realized Volatility", fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Histogram of Realized Volatility
plt.figure(figsize=(12, 5))
sns.histplot(asml_df["realized_volatility"], bins=10, label="ASML", kde=True, alpha=0.6)
sns.histplot(bmw_df["realized_volatility"], bins=10, label="BMW", kde=True, alpha=0.6)

plt.title("Distribution of Realized Volatility", fontsize=14)
plt.xlabel("Realized Volatility", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.show()

# Boxplot comparison
plt.figure(figsize=(8, 5))
sns.boxplot(data=[asml_df["realized_volatility"], bmw_df["realized_volatility"]], palette="Set2")
plt.xticks([0, 1], ["ASML", "BMW"])
plt.title("Boxplot of Realized Volatility", fontsize=14)
plt.ylabel("Realized Volatility", fontsize=12)
plt.show()

# Display summary statistics
print("\n--- ASML Realized Volatility Statistics ---")
print(asml_df["realized_volatility"].describe())

print("\n--- BMW Realized Volatility Statistics ---")
print(bmw_df["realized_volatility"].describe())

```
