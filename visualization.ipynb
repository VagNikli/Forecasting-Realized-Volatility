```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the realized volatility data
asml_vol = pd.read_csv(r'C:\Users\vagel\Desktop\Forecasting Realized Volatility\asml_realized_volatility.csv')
bmw_vol = pd.read_csv(r'C:\Users\vagel\Desktop\Forecasting Realized Volatility\bmw_realized_volatility.csv')

# Convert date column to datetime
asml_vol['date'] = pd.to_datetime(asml_vol['date'])
bmw_vol['date'] = pd.to_datetime(bmw_vol['date'])

# Merge datasets for correlation analysis
merged_vol = pd.merge(asml_vol, bmw_vol, on="date", suffixes=('_asml', '_bmw'))
```

```python
# Time Series of Realized Volatility
plt.figure(figsize=(10, 6))
plt.plot(asml_vol['date'], asml_vol['realized_volatility'], label='ASML', color='blue')
plt.plot(bmw_vol['date'], bmw_vol['realized_volatility'], label='BMW', color='red')
plt.xlabel('Date')
plt.ylabel('Realized Volatility')
plt.title('Time Series of Realized Volatility for ASML and BMW')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

### 1. Time Series of Realized Volatility
**Depicts:** The daily realized volatility trends for ASML and BMW stocks over time.

**Interpretation:** Periods of higher volatility indicate increased market uncertainty, while stable periods suggest low price fluctuations. Comparing the two stocks helps assess their volatility dynamics and any similarities in their behavior.


```python
# Histogram of Realized Volatility
plt.figure(figsize=(12, 6))
sns.histplot(asml_vol['realized_volatility'], kde=True, bins=30, label='ASML', color='blue', alpha=0.6)
sns.histplot(bmw_vol['realized_volatility'], kde=True, bins=30, label='BMW', color='red', alpha=0.6)
plt.xlabel('Realized Volatility')
plt.ylabel('Frequency')
plt.title('Histogram of Realized Volatility for ASML and BMW')
plt.legend()
plt.show()
```

### 2. Histogram of Realized Volatility
**Depicts:** The frequency distribution of realized volatility values for ASML and BMW.

**Interpretation:** Peaks in the histogram indicate the most common volatility levels, while the spread suggests how volatile each stock tends to be. Wider distributions imply greater variability in realized volatility.


```python
# Box Plot of Realized Volatility
plt.figure(figsize=(8, 6))
sns.boxplot(data=[asml_vol['realized_volatility'], bmw_vol['realized_volatility']], palette=['blue', 'red'])
plt.xticks([0, 1], ['ASML', 'BMW'])
plt.ylabel('Realized Volatility')
plt.title('Box Plot of Realized Volatility for ASML and BMW')
plt.show()
```

### 3. Box Plot of Realized Volatility
**Depicts:** The realized volatility distributions of ASML and BMW.

**Interpretation:** The median line within each box represents the central tendency, while the whiskers indicate the range of volatility. Outliers, if present, suggest extreme volatility events. This visualization helps compare the overall stability of both stocks.


```python
# Correlation between ASML and BMW Realized Volatility
plt.figure(figsize=(8, 6))
sns.scatterplot(x=merged_vol['realized_volatility_asml'], y=merged_vol['realized_volatility_bmw'])
plt.xlabel('ASML Realized Volatility')
plt.ylabel('BMW Realized Volatility')
plt.title('Correlation between ASML and BMW Realized Volatility')
plt.show()
```

### 4. Correlation between ASML and BMW Realized Volatility
**Depicts:** The relationship between ASML and BMW realized volatilities.

**Interpretation:** A strong positive correlation would suggest that the volatilities of the two stocks move together, likely influenced by broader market conditions. A weak correlation, on the other hand, indicates independent volatility patterns.
